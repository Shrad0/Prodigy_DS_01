```python
year = 2012
variable = 'Life expectancy at birth (years)'
```

```python
data_year = data[data['Year'] == year]
```

```python
#Histogram

plt.figure(figsize=(10,6))
plt.hist(data_year[variable], bins=20, color='g', edgecolor='k', alpha=0.7)
plt.title(f'Distribution of {variable} in {year}')
plt.xlabel(variable)
plt.ylabel('Frequency')
plt.show()
```

```python
plt.title(f'Distribution of {variable} in {year}')
plt.xlabel(variable)
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)

mean_value = data_year[variable].mean()
median_value = data_year[variable].median()
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='blue', linestyle='dashed', linewidth=2, label=f'Median: {median_value:.2f}')
plt.legend()
plt.show()
```

```python
region_counts = data['Region'].value_counts()

#Bar Chart
plt.figure(figsize=(10,6))
region_counts.plot(kind = 'bar', color='magenta', edgecolor='k')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.title('Distribution of countries by region')
plt.xticks(rotation = 45)
plt.show()
```

```python
#Pie Chart
plt.figure(figsize=(15,6))
plt.subplot(1,2,2)
region_percentage = (region_counts / region_counts.sum()) * 100
colors = ['gold', 'lightgreen', 'lightcoral', 'lightskyblue']
region_percentage.plot(kind='pie', autopct='%1.1f%%', colors=colors)
plt.axis('equal')
plt.title('Distribution of countries by region (percentage)')
plt.tight_layout()
plt.show()
```
